```python
sales_df['Address'] = sales_df['Address'].str.replace(r",(\S)", r", \1", regex=True)
```

```python
# Modify the date columns to datetime data type to ensure consistency
```

```python
sales_df['Order Date'] = pd.to_datetime(sales_df['Order Date'], dayfirst = True)
```

Some numerical columns have an object datatype. To better perform our analysis, let's modify "cost" related columns to a float datatype. First, I will remove the dollar signs and commas then convert.

sales_df['Cost Price'] = sales_df['Cost Price'].replace({'\$': ''}, regex=True).astype(float)

```python
sales_df['Retail Price'] = sales_df['Retail Price'].replace({'\$': ''}, regex=True).astype(float)
```

```python
sales_df['Profit Margin'] = sales_df['Profit Margin'].replace({'\$': ''}, regex=True).astype(float)
```

```python
sales_df['Sub Total'] = sales_df['Sub Total'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
sales_df['Discount %'] = sales_df['Discount %'].replace({'\%': ''}, regex=True).astype(float)
```

```python
sales_df['Discount $'] = sales_df['Discount $'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
sales_df['Order Total'] = sales_df['Order Total'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
sales_df['Shipping Cost'] = sales_df['Shipping Cost'].replace({'\$': ''}, regex=True).astype(float)
```

```python
sales_df['Total'] = sales_df['Total'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
# Specify the columns to check for outliers
```

```python
columns_to_check = ['Cost Price', 'Retail Price', 'Profit Margin', 'Sub Total', 'Discount $', 'Order Total', 'Total']
```

```python
# Calculate Q1, Q3, and IQR for the specified columns
```

```python
Q1 = sales_df[columns_to_check].quantile(0.25)
```

```python
Q3 = sales_df[columns_to_check].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
# Define outliers bound
```

```python
lower_bound = Q1 - 1.5 * IQR
```

```python
upper_bound = Q3 + 1.5 * IQR
```

```python
# Filter out outliers
```

```python
filtered_sales = sales_df[~((sales_df[columns_to_check] < lower_bound) | (sales_df[columns_to_check] > upper_bound)).any(axis=1)].copy()
```

```python
filtered_sales.head(10)
```

```python
# Sort the dataset by the Order Date column 
```

```python
filtered_sales.sort_values(by= ['Order Date'], inplace= True)
```

```python
filtered_sales
```

```python
# Reset the index
```

```python
filtered_sales.reset_index(drop= True, inplace= True)
```

```python
filtered_sales
```

Let's create a new column to store the cumulative revenue that will be used in visualization

```python
filtered_sales['Cumulative Revenue'] = filtered_sales['Total'].cumsum()
```
